Schritt 1: Importieren Sie die benötigten Bibliotheken

```python
import numpy as np
import scipy as sp
import scipy.signal as sg
import matplotlib.pyplot as plt
import imageio.v2 as imageio
```

Schritt 2: Lesen Sie das Bild 'lenna_noise.png' mit Hilfe der imageio-Bibliothek ein!

```python
image = imageio.imread("data/lenna_noise.png")
image.shape
```

Schritt 3: Geben Sie das Bild mit matplotlib aus

```python
plt.imshow(image, cmap="gray")
plt.show()
```

Schritt 4: Filtern Sie das Bild mit einem 3x3 Mittelwertfilter wie in Folie 23 gezeigt und plotten Sie das Ergebnis.

```python
avg_filter = np.array([[1,1,1], [1,1,1], [1,1,1]])
avg_filter
```

```python
image_avg_filtered = sg.convolve(image, avg_filter) / 9
plt.imshow(image_avg_filtered, cmap="gray")
plt.show()
```

Schritt 5: Filtern Sie das Bild mit einem 3x3 Kantenfilter wie in Folie 26 gezeigt und plotten Sie das Ergebnis.

```python
edge_filter = np.array([[-1,1,0], [-1,1,0], [-1,1,0]])
edge_filter
```

```python
image_edge_filtered = sg.convolve(image, edge_filter)
plt.imshow(image_edge_filtered, cmap="gray")
plt.show()
```
